## Objective  

The objective of the Netflix Data Analytics project is to perform a comprehensive analysis of the content available on Netflix to uncover valuable insights. This involves analyzing the distribution of content types (movies vs. TV shows), ratings, and genres, as well as investigating the geographical distribution of content across different countries. The findings will be visually represented using charts, graphs, and heatmaps to effectively communicate the insights. Overall, the project aims to provide a detailed understanding of Netflix's content landscape, aiding content creators, marketers, and viewers in making informed decisions.

### 2. In which year highest number of the TV shows & Movies were released?  

```python
data.dtypes 
```

```python
# Convert 'date_added' column to datetime format
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
```

```python
# Group by release year and count the number of releases
release_counts = data.groupby('release_year').size().reset_index(name='count')

# Find the year with the highest number of releases
max_releases_year = release_counts[release_counts['count'] == release_counts['count'].max()]

print(max_releases_year)
```

```python
# Print the year with the highest number of releases
print(f"Year with the highest number of releases: {max_releases_year['release_year'].values[0]}")
print(max_releases_year)
```

```python
# Plot the number of releases per year
release_counts.plot(kind='bar', x='release_year', y='count', figsize=(12, 6), legend=False)
plt.title('Number of Releases by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.show()
```

```python
# Filter data for the years 2012 to 2024
filtered_data = data[(data['release_year'] >= 2012) & (data['release_year'] <= 2024)]

# Group by release year and count the number of releases
release_counts = filtered_data.groupby('release_year').size().reset_index(name='count')

# Find the year with the highest number of releases
max_releases_year = release_counts[release_counts['count'] == release_counts['count'].max()]
```

```python
# Plot the number of releases per year from 2012 to 2024
release_counts.plot(kind='bar', x='release_year', y='count', figsize=(12, 6), legend=False)
plt.title('Number of Releases by Year (2012-2024)')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.show()
```

## Advance Analysis  

### 1. Sentiment Analysis on Descriptions 

```python
from textblob import TextBlob
```

```python
# Perform sentiment analysis on the descriptions
filtered_data['description_sentiment'] = filtered_data['description'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Plot the distribution of sentiment scores
plt.figure(figsize=(10, 6))
sns.histplot(filtered_data['description_sentiment'], bins=20, kde=True)
plt.title('Distribution of Description Sentiment (2012-2024)')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

### 2. Genre Evolution Over Time

```python
# Extract genres and release years
genres = filtered_data['listed_in'].str.split(', ').explode().reset_index()
genres['release_year'] = filtered_data.loc[genres['index'], 'release_year'].values
```

```python
# Count the number of shows/movies per genre per year
genre_trend = genres.groupby(['release_year', 'listed_in']).size().unstack().fillna(0)
```

```python
# Plot the trends of top 5 genres over time
top_5_genres = genre_trend.sum().sort_values(ascending=False).head(5).index
genre_trend[top_5_genres].plot(figsize=(14, 7))
plt.title('Trend of Top 5 Genres (2012-2024)')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.show()
```

### 3. Geographical Analysis 

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Load world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
# Extract and count content by country
country_count = filtered_data['country'].str.split(', ').explode().value_counts()
```

```python
# Create a DataFrame with counts and merge with world map
country_df = pd.DataFrame(country_count).reset_index()
country_df.columns = ['country', 'count']
world = world.merge(country_df, how='left', left_on='name', right_on='country').fillna(0)
```

```python
# Plot the geographical distribution
world.plot(column='count', cmap='OrRd', legend=True, figsize=(14, 7), legend_kwds={'label': "Number of Releases"})
plt.title('Geographical Distribution of Content (2012-2024)')
plt.show()
```

### 4. Rating Analysis

```python
# Plot the distribution of ratings
plt.figure(figsize=(10, 6))
sns.countplot(y=filtered_data['rating'], order=filtered_data['rating'].value_counts().index)
plt.title('Distribution of Ratings (2012-2024)')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

```python
# Trend of ratings over time
rating_trend = filtered_data.groupby(['release_year', 'rating']).size().unstack().fillna(0)
rating_trend.plot(figsize=(14, 7), cmap='tab20', linewidth=2)
plt.title('Trend of Ratings Over Time (2012-2024)')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.show()
```
